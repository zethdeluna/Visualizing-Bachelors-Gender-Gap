# Visualizing the Gender Gap in College Degrees

In this project, we are going to visualize the change in the gender gap for students that achieved a bachelor's degree in the given majors. The process for cleaning and compiling data, as well as the final "percent_female_bachelors.csv" dataset, are available [here](https://github.com/zethdeluna/Bachelor-s-Degrees-Female-Percentage).

The data set only has information on the yearly percentage of female students in each major, so we'll have to do a simple calculation to find the amount of male students per major. Once we have both female and male percentages, we can compare the change in gender gap per major from 1971 to 2018.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degree = pd.read_csv('percent_female_bachelors.csv', header=0)
```

```python
women_degree.head()
```

First, let's group the majors by broader field of study.

```python
stem = ['biology',
        'computer science',
        'engineering',
        'math and statistics',
        'physical sciences',
        'psychology']

lib_arts = ['art and performance',
            'communications and journalism',
            'english',
            'foreign languages',
            'social sciences and history']

other = ['agriculture',
         'architecture',
         'business',
         'education',
         'health professions',
         'public administration']
```

Let's first order each group by descending percentage of female graduates per major in 2018. Then, we'll plot the change in gender gap for each major group and display them in a grid. 

__Organizing STEM Majors by Descending Percentage of Female Graduates in 2018__

```python
# create dictionary that has stem majors as keys
# and percent of female graduates in 2018 as values
stem_dict_2018 = {}
for major in stem:
    stem_dict_2018[major] = women_degree.loc[47, major]
    
# create a list of stem majors, sorted by descending 
# percentage of female graduates in 2018
stem_2018 = sorted(stem_dict_2018, key=stem_dict_2018.get, reverse=True)
```

```python
stem_dict_2018
```

```python
stem_2018
```

__Organizing Liberal Arts Majors by Descending Percentage of Female Graduates in 2018__

```python
# similar process as sorting the STEM list
libarts_dict_2018 = {}
for major in lib_arts:
    libarts_dict_2018[major] = women_degree.loc[47, major]
    
libarts_2018 = sorted(libarts_dict_2018, key=libarts_dict_2018.get, reverse=True)
```

```python
libarts_dict_2018
```

```python
libarts_2018
```

__Organizing "Other" Majors by Descending Percentage of Female Graduates in 2018__

```python
# similar process as the two above
other_dict_2018 = {}
for major in other:
    other_dict_2018[major] = women_degree.loc[47, major]
    
other_2018 = sorted(other_dict_2018, key=other_dict_2018.get, reverse=True)
```

## Gender Gap Line Charts

Now that we have the majors in the order that we want, let's visualize the data using line graphs.

```python
# Lakers Colors
purple = (85/255, 37/255, 130/255)
gold = (253/255, 185/255, 39/255)
black = (6/255, 25/255, 34/255)
```

```python
fig = plt.figure(figsize=(24, 24), dpi=500)
fig.suptitle('Change in Gender Gaps within Majors', fontsize=30)

# create STEM majors line graphs
for i in range(0,18,3):
    major = int(i/3)
    x = women_degree['year']
    y_women = women_degree[stem_2018[major]]
    y_men = 100 - y_women
    ax = fig.add_subplot(6, 3, i+1)
    ax.set_title(stem_2018[major])
    ax.plot(x, y_women, color=purple, linewidth=3)
    ax.plot(x, y_men, color=gold, linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top=False,
                   bottom=False,
                   left=False,
                   right=False,
                   labelbottom=False)
    ax.set_xlim(1970, 2020)
    ax.set_ylim(0, 100)
    ax.set_yticks([0,100])
    ax.axhline(50, color=black, alpha=0.4)

    if major == 0:
        ax.text(2014, 85, 'Women')
        ax.text(2014, 14, 'Men')
    elif major == 5:
        ax.text(2014, 9, 'Women')
        ax.text(2014, 88, 'Men')
        ax.tick_params(labelbottom=True)
    
# create Liberal Arts majors line graphs
for i in range(0,15,3):
    major = int(i/3)
    x = women_degree['year']
    y_women = women_degree[libarts_2018[major]]
    y_men = 100 - y_women
    ax = fig.add_subplot(6, 3, i+2)
    ax.set_title(libarts_2018[major])
    ax.plot(x, y_women, color=purple, linewidth=3)
    ax.plot(x, y_men, color=gold, linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top=False,
                   bottom=False,
                   left=False,
                   right=False,
                   labelbottom=False)
    ax.set_xlim(1970, 2020)
    ax.set_ylim(0, 100)
    ax.set_yticks([0,100])
    ax.axhline(50, color=black, alpha=0.4)
    
    if major == 0:
        ax.text(2014, 77, 'Women')
        ax.text(2014, 22, 'Men')
    elif major == 4:
        ax.tick_params(labelbottom=True)
    
    
# create Other majors line graphs
for i in range(0,18, 3):
    major = int(i/3)
    x = women_degree['year']
    y_women = women_degree[other_2018[major]]
    y_men = 100 - y_women
    ax = fig.add_subplot(6, 3, i+3)
    ax.set_title(other_2018[major])
    ax.plot(x, y_women, color=purple, linewidth=3)
    ax.plot(x, y_men, color=gold, linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top=False,
                   bottom=False,
                   left=False,
                   right=False,
                   labelbottom=False)
    ax.set_xlim(1970, 2020)
    ax.set_ylim(0, 100)
    ax.set_yticks([0,100])
    ax.axhline(50, color=black, alpha=0.4)
    
    if major == 0:
        ax.text(2014, 90, 'Women')
        ax.text(2014, 8, 'Men')
    elif major == 5:
        ax.text(2014, 38, 'Women')
        ax.text(2014, 57, 'Men')
        ax.tick_params(labelbottom=True)

plt.savefig('gender_gap_2018.png')
plt.show()
```
